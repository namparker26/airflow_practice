Read the data from database via a Spark Connection

```python
soccer_data_objects = [
    'Country',
    'League',
    'Match',
    'Player',
    'Player_Attributes',
    'Team',
    'Team_Attributes']

data_path = 'data'
cwd = f"{os.getcwd()}/ddl"
for data_object in soccer_data_objects:
    obj = SqliteDataObject(data_object)
    obj.write_to_parquet()

    sourcing_object = SourcingTable(data_path, data_object, 'sourcing', data_object)
    object_ddl_query_statement = sourcing_object.generate_ddl_query()
    notebook = Notebook(cwd, data_object)
    notebook.create_notebook()
    notebook.edit_cell_content([object_ddl_query_statement], '%sql')
    notebook.inspect_notebook_cell()
```
