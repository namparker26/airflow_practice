Read the data from database via a Spark Connection

```python
db_path = "/Users/namnguyen/Documents/Documents on MacPro/Projects/Airflows/database.sqlite"
spark = (
    SparkSession.builder
    .appName("SoccerSQLite")
    .config("spark.jars.packages", "org.xerial:sqlite-jdbc:3.48.0.0")
    .getOrCreate()
) 
jdbc_url = f"jdbc:sqlite:{db_path}"

def query_from_sqllite_and_write(table_name):
    df = (
        spark.read.format("jdbc")
        .option("url", jdbc_url)
        .option("dbtable", table_name)
        .option("driver", "org.sqlite.JDBC")
        .load()
    )

    output_path = f'/Users/namnguyen/Documents/Documents on MacPro/Projects/Airflows/data/{table_name}'
    df.write.mode("overwrite").parquet(output_path)
```

```python
query_from_sqllite_and_write('Match')
```
